# General Forward Propagation

In this notebook we extend a single neuron into a layer of multiple neurons.

Objectives:

- Understand vectorized computation
- Understand matrix multiplication
- Build a neural network layer
- Visualize neuron outputs
- Prepare for TensorFlow implementation

```python
import numpy as np
X=np.array([10,2])
W = np.array([
    [0.5, 1.0, -0.5],
    [0.8, 0.2, 1.5]
])
B=np.array([1,2,0.5])
Z=np.dot(X,W)+B
print(Z)
```

```python
def relu(x):
    return np.maximum(0,x)
A=relu(Z)
print("after applying ReLu activation function:",A)
```

```python
print("Input shape:",X.shape)
print("Weights shape:",W.shape)
print("BIas shape:",B.shape)
print("Output shape:",A.shape)
```

2 inputs

3 neurons

therefore weight matrix = 2 × 3

```python
contributions=X.reshape(-1,1)*W
print(contributions)
```

```python
import matplotlib.pyplot as plt
neurons=["Neuron1","Neuron2","Neuron3"]
plt.figure(figsize=(8,6))
plt.bar(neurons,A)
plt.title("Output of Neurons after ReLU Activation")
plt.xlabel("Neurons")   
plt.ylabel("Output Value")
plt.show()


```
