```python
import tensorflow as tf
layer=tf.keras.layers.Dense(units=5,activation='relu')
print(layer)
```

```python
import numpy as np
X = np.array([[20,170,75,7,5,1,3,4]])
print(X.shape)
```

```python
output=layer(X)
print(output)
```

```python
import matplotlib.pyplot as plt
outputs=output.numpy()[0]
plt.bar(range(len(outputs)),outputs)
plt.title("Output of the Neural Network Layer")
plt.xlabel("Neuron index")
plt.ylabel("Activation value")
plt.show()
```

```python
weights, biases=layer.get_weights()
print("Weights:\n",weights) 
print("Biases:\n",biases)
print("Weights shape:",weights.shape)
```

```python
plt.imshow(weights, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title("Visualization of Weights")
plt.xlabel("Output Neuron Index")
plt.ylabel("Input Feature Index")
plt.show()  
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(2)
])
prediction=model(X)
print("Prediction:\n",prediction)
```
