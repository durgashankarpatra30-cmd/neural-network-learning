# Single Layer Forward Propagation

This notebook demonstrates how a single neuron performs forward propagation.

## Objectives

- Understand inputs
- Understand weights
- Understand bias
- Calculate weighted sum
- Apply activation functions

```python
import numpy as np
#input
x1=5;
x2=8;
#weights
w1=0.5;
w2=0.8;
#bias
b=1;
print("x1=",x1,"x2=",x2,"w1=",w1,"w2=",w2,"b=",b);
```

```python
z=w1*x1+w2*x2;
print(z);
```

```python
def relu(x):
    return np.maximum(0,x)
a=relu(z);
print("After applying ReLU activation function:",a);
```

```python
test_values = [-5, -2, 0, 3, 7]

for value in test_values:
    print(f"ReLU({value}) = {relu(value)}")

```
